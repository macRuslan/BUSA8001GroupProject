**Evaluation Criteria:** 

In this project, we will employ **RMSE** to evaluate the performance of the ***Ridge Regression model***, ***Lasso Regression model*** and ***Decision Tree***.

Root-mean-square error (RMSE) is a measure used to quantify the difference between a set of predicted values and the corresponding observed values. In machine learning, RMSE provides an advantage as it gives a clear sense of the average magnitude of prediction errors, making it a valuable tool for assessing the accuracy and reliability of regression models.


$$
RMSE = \sqrt{\frac{1}{N} \sum_{i=1}^{N} (y_i - \hat{y}_i)^2}
$$

Where:
- \(N\) is the total number of observations.
- \(y_i\) is the observed value for the i-th observation.
- \(\hat{y}_i\) is the predicted value for the i-th observation.

```python
# Check the number of missing values in the train and test datasets
missing_train = train_data.isnull().sum()
missing_test = test_data.isnull().sum()

# Combine the missing values data for a consolidated view
missing_values = pd.DataFrame({'Train Missing': missing_train, 'Test Missing': missing_test})
missing_values[missing_values.sum(axis=1) > 0]
```

The table above displays the number of missing values in each column for both the training and test datasets.

For imputation:

- Numerical features will be imputed with their median value.

- Categorical features will be imputed with the most frequent value (mode).

```python
# List of columns for imputation
columns_to_impute = missing_values.index.tolist()

# Impute missing values for both train and test datasets

for column in columns_to_impute:
    if column in train_data.columns and column in test_data.columns:
        if train_data[column].dtype in ['float64', 'int64']:  # Numerical feature
            median_value = train_data[column].median()  # Calculate median from training data
            train_data[column].fillna(median_value, inplace=True)
            test_data[column].fillna(median_value, inplace=True)
        else:  # Categorical feature
            mode_value = train_data[column].mode()[0]  # Calculate mode from training data
            train_data[column].fillna(mode_value, inplace=True)
            test_data[column].fillna(mode_value, inplace=True)

# Verify if there are any missing values remaining
missing_train_after = train_data.isnull().sum().sum()
missing_test_after = test_data.isnull().sum().sum()

missing_train_after, missing_test_after
```

```python
# Remove 'vin' from the list of columns to one-hot encode
categorical_columns.remove('vin')
categorical_columns
```

```python
def process_categorical_column(series):
    # Mark observations with multiple values as 'other'
    series[series.str.contains(',', na=False)] = 'other'
    
    # For columns with more than 5 unique values, retain the 5 most frequent ones and map others to 'other'
    if series.nunique() > 5:
        top_5_values = series.value_counts().head(5).index.tolist()
        series[~series.isin(top_5_values)] = 'other'
    
    return series

# Process the categorical columns in the training dataset
for column in categorical_columns:
    train_data[column] = process_categorical_column(train_data[column].copy())

# Process the categorical columns in the testing dataset
for column in categorical_columns:
    if column in test_data.columns:
        test_data[column] = process_categorical_column(test_data[column].copy())
        
# Copy the original train and test datasets to new dataframes for one-hot encoding
train_data_encoded = train_data.copy()
test_data_encoded = test_data.copy()

# Perform one-hot encoding for the processed categorical columns in the training dataset
train_encoded = pd.get_dummies(train_data, columns=categorical_columns, drop_first=True)

# Perform one-hot encoding for the processed categorical columns in the test dataset
test_encoded = pd.get_dummies(test_data, columns=[col for col in categorical_columns if col in test_data.columns], drop_first=True)

# Display the first few rows of the encoded training dataset
train_encoded.head()
```

```python
from scipy.stats import ttest_ind

# Example: T-test between prices of new and used cars
new_car_prices = train_encoded[train_encoded['is_new'] == True]['price']
used_car_prices = train_encoded[train_encoded['is_new'] == False]['price']

t_stat, p_value = ttest_ind(new_car_prices, used_car_prices)
print('T-statistic:', t_stat)
print('P-value:', p_value)
```

**T-statistic (30.9917):**

This value represents the difference between the means of the two groups (new car prices and used car prices) in terms of standard errors. A T-statistic value far from 0 suggests that the group means are different.
The positive value of the T-statistic indicates that the mean price of the new cars is greater than the mean price of the used cars.

**P-value (1.495753847717702e-186):**

The P-value represents the probability of observing a T-statistic as extreme as the one calculated, assuming that there is no actual difference between the means of the two groups (null hypothesis).
A very small P-value, like the one provided, indicates strong evidence against the null hypothesis. Given such a small P-value, one would typically reject the null hypothesis and conclude that there is a significant difference in the mean prices of new and used cars.
Conclusion:
The T-test results suggest that there is a statistically significant difference between the prices of new and used cars in the dataset. Given the positive T-statistic and the extremely low P-value, it can be inferred that new cars are priced higher than used cars in this dataset.

```python
# Get the first feature from each pair
features_to_drop_train = [
    'city_fuel_economy',
    'engine_displacement',
    'highway_fuel_economy',
    'length',
    'wheelbase',
    'engine_displacement_liters',
    'engine_type_I4 Hybrid',
    'fuel_type_Hybrid',
    'transmission_CVT',
    'transmission_display_Continuously Variable Transmission'
]

features_to_drop_test = [
    'city_fuel_economy',
    'engine_displacement',
    'highway_fuel_economy',
    'length',
    'wheelbase',
    'engine_displacement_liters',
    'fuel_type_Hybrid',
    'transmission_CVT'
]

# Drop these features from the train and test dataframes
train_encoded_without_vim = train_encoded_without_vim.drop(columns=features_to_drop_train)
test_encoded_without_vim = test_encoded_without_vim.drop(columns=features_to_drop_test)
```

```python
# Let's check if the train and test datasets have the same columns
# Get the column names of the training dataset
train_columns = list(train_encoded_without_vim.columns)

# Remove the column "price" from the list of train columns
train_columns.remove("price")

# Get the column names of the testing dataset
test_columns = list(test_encoded_without_vim.columns)

# Check if all columns in the training set are present in the testing set
if all(column in test_columns for column in train_columns):
    print("All columns in the training set are present in the testing set.")
    # All columns in the training set are present in the testing set
else:
    # Get the list of columns in the training set that are not present in the testing set
    drop_columns = list(set(train_columns) - set(test_columns))

    # Ensure the 'price' feature is not in drop_columns
    if 'price' in drop_columns:
        drop_columns.remove('price')

    # Drop the columns from the training dataset
    train_encoded_without_vim.drop(
        columns=drop_columns,
        inplace=True
    )

# Check if all columns in the testing set are present in the training set
if all(column in train_columns for column in test_columns):
    print("All columns in the testing set are present in the training set.")
    # All columns in the testing set are present in the training set
else:
    # Get the list of columns in the testing set that are not present in the training set
    drop_columns = list(set(test_columns) - set(train_columns))

    # Drop the columns from the testing dataset
    test_encoded_without_vim.drop(
        columns=drop_columns,
        inplace=True
    )

train_columns = list(train_encoded_without_vim.columns)

# Remove the column "price" from the list of train columns
train_columns.remove("price")

# Get the column names of the testing dataset
test_columns = list(test_encoded_without_vim.columns)

# Check if the two sets of column names are equal
if train_columns == test_columns:
    print("The training and testing datasets share the same columns.")
else:
    print("The training and testing datasets do not share the same columns.")
```

```python
# We started with Ridge Regression model
from skopt import BayesSearchCV
from sklearn.linear_model import Ridge
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import mean_squared_error
from sklearn.feature_selection import SelectKBest, f_regression
import numpy as np
import joblib

# Split the dataframe into features (X) and target variable (y)
X = train_encoded_without_vim.drop('price', axis=1)
y = train_encoded_without_vim['price']

# Define the search space for alpha hyperparameter
search_space = [{'alpha': (0.001, 10.0, 'log-uniform')}]

# Create a Ridge regression model
ridge = Ridge()

# Perform hyperparameter tuning with Bayesian optimization for Ridge model
ridge_opt = BayesSearchCV(ridge, search_space, n_iter=30, cv=5, n_jobs=-1)

# Fit the Ridge model with optimized hyperparameters to the training data
ridge_opt.fit(X, y)

# Print the best alpha value for Ridge model
print("Best alpha for Ridge: ", ridge_opt.best_params_['alpha'])

# Save the Ridge model
joblib.dump(ridge_opt, 'ridge_model.pkl')

# Load the saved Ridge model
ridge_model = joblib.load('ridge_model.pkl')

# Perform cross-validation and make predictions on the training data
ridge_predictions = cross_val_predict(ridge_model, X, y, cv=5)

# Apply inverse transformation to the predictions (since log transformation was applied to the target variable)
ridge_predictions = np.exp(ridge_predictions)

# Calculate RMSE on the training data
rmse = np.sqrt(mean_squared_error(np.exp(y), ridge_predictions))
print("RMSE on training data: ", rmse)
```

```python
# We want to check which features are most important for the Ridge model
# Perform feature selection using SelectKBest with f_regression scoring
selector = SelectKBest(score_func=f_regression, k=40)
selected_features = selector.fit_transform(X, y)

# Get the selected feature indices
selected_indices = selector.get_support(indices=True)

# Get the names of the selected features
ridge_selected_feature_names = X.columns[selected_indices]

# Print the selected feature names
print("Selected Features: ", ridge_selected_feature_names)
```
