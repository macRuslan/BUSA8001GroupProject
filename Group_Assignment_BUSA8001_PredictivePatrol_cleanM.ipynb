**Evaluation Criteria:** In this project, we will employ RMSD to evaluate the performance of the ***Model which we will use in the next tasks***.

Root-mean-square deviation (RMSD) is a measure used to quantify the difference between two sets of data points, typically representing predicted values and observed values. In machine learning, RMSD provides an advantage as it gives a clear sense of the average magnitude of prediction errors, making it a valuable tool for assessing the accuracy and reliability of regression models.

$$
RMSD = \sqrt{\frac{1}{N} \sum_{i=1}^{N} (y_i - \hat{y}_i)^2}
$$
Where:
- $N$ is the total number of observations.
- $y_i$ is the observed value for the i-th observation.
- $\hat{y}_i$ is the predicted value for the i-th observation.

```python
# Check the number of missing values in the train and test datasets
missing_train = train_data.isnull().sum()
missing_test = test_data.isnull().sum()

# Combine the missing values data for a consolidated view
missing_values = pd.DataFrame({'Train Missing': missing_train, 'Test Missing': missing_test})
missing_values[missing_values.sum(axis=1) > 0]
```

The table above displays the number of missing values in each column for both the training and test datasets.

For imputation:

- Numerical features will be imputed with their median value.

- Categorical features will be imputed with the most frequent value (mode).

```python
# List of columns for imputation
columns_to_impute = missing_values.index.tolist()

# Impute missing values for both train and test datasets

for column in columns_to_impute:
    if column in train_data.columns and column in test_data.columns:
        if train_data[column].dtype in ['float64', 'int64']:  # Numerical feature
            median_value = train_data[column].median()  # Calculate median from training data
            train_data[column].fillna(median_value, inplace=True)
            test_data[column].fillna(median_value, inplace=True)
        else:  # Categorical feature
            mode_value = train_data[column].mode()[0]  # Calculate mode from training data
            train_data[column].fillna(mode_value, inplace=True)
            test_data[column].fillna(mode_value, inplace=True)

# Verify if there are any missing values remaining
missing_train_after = train_data.isnull().sum().sum()
missing_test_after = test_data.isnull().sum().sum()

missing_train_after, missing_test_after
```

```python
# Identify categorical columns in the training dataset
categorical_columns = train_data.select_dtypes(include=['object']).columns.tolist()
categorical_columns
```

We have identified the following categorical columns: `vin`, `body_type`, `city`, `engine_type`, `exterior_color`, `fuel_type`, `interior_color`, `listed_date`, `listing_color`, `make_name`, `maximum_seating`, `model_name`, `power`, `torque`, `transmission`, `transmission_display`, and `wheel_system`.

```python
# Remove 'vin' from the list of columns to one-hot encode
categorical_columns.remove('vin')
categorical_columns
```

```python
def process_categorical_column(series):
    # Mark observations with multiple values as 'other'
    series[series.str.contains(',', na=False)] = 'other'
    
    # For columns with more than 5 unique values, retain the 5 most frequent ones and map others to 'other'
    if series.nunique() > 5:
        top_5_values = series.value_counts().head(5).index.tolist()
        series[~series.isin(top_5_values)] = 'other'
    
    return series

# Process the categorical columns in the training dataset
for column in categorical_columns:
    train_data[column] = process_categorical_column(train_data[column].copy())

# Process the categorical columns in the testing dataset
for column in categorical_columns:
    if column in test_data.columns:
        test_data[column] = process_categorical_column(test_data[column].copy())
        
# Copy the original train and test datasets to new dataframes for one-hot encoding
train_data_encoded = train_data.copy()
test_data_encoded = test_data.copy()

# Perform one-hot encoding for the processed categorical columns in the training dataset
train_encoded = pd.get_dummies(train_data, columns=categorical_columns, drop_first=True)

# Perform one-hot encoding for the processed categorical columns in the test dataset
test_encoded = pd.get_dummies(test_data, columns=[col for col in categorical_columns if col in test_data.columns], drop_first=True)

# Display the first few rows of the encoded training dataset
train_encoded.head()
```

```python
# We started with Ridge Regression model
from skopt import BayesSearchCV
from sklearn.linear_model import Ridge
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import mean_squared_error
from sklearn.feature_selection import SelectKBest, f_regression
import numpy as np
import joblib

# Split the dataframe into features (X) and target variable (y)
X = train_encoded_without_vim.drop('price', axis=1)
y = train_encoded_without_vim['price']

# Define the search space for alpha hyperparameter
search_space = [{'alpha': (0.001, 10.0, 'log-uniform')}]

# Create a Ridge regression model
ridge = Ridge()

# Perform hyperparameter tuning with Bayesian optimization for Ridge model
ridge_opt = BayesSearchCV(ridge, search_space, n_iter=30, cv=5, n_jobs=-1)

# Fit the Ridge model with optimized hyperparameters to the training data
ridge_opt.fit(X, y)

# Print the best alpha value for Ridge model
print("Best alpha for Ridge: ", ridge_opt.best_params_['alpha'])

# Save the Ridge model
joblib.dump(ridge_opt, 'ridge_model.pkl')

# Load the saved Ridge model
ridge_model = joblib.load('ridge_model.pkl')

# Perform cross-validation and make predictions on the training data
ridge_predictions = cross_val_predict(ridge_model, X, y, cv=5)

# Apply inverse transformation to the predictions (since log transformation was applied to the target variable)
ridge_predictions = np.exp(ridge_predictions)

# Calculate RMSE on the training data
rmse = np.sqrt(mean_squared_error(np.exp(y), ridge_predictions))
print("RMSE on training data: ", rmse)
```

```python
# We want to check which features are most important for the Ridge model
# Perform feature selection using SelectKBest with f_regression scoring
selector = SelectKBest(score_func=f_regression, k=10)
selected_features = selector.fit_transform(X, y)

# Get the selected feature indices
selected_indices = selector.get_support(indices=True)

# Get the names of the selected features
ridge_selected_feature_names = X.columns[selected_indices]

# Print the selected feature names
print("Selected Features: ", ridge_selected_feature_names)
```
