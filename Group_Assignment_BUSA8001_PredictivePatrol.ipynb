**Evaluation Criteria:** In this project, we will employ RMSD to evaluate the performance of the ***Model which we will use in the next tasks***.

Root-mean-square deviation (RMSD) is a measure used to quantify the difference between two sets of data points, typically representing predicted values and observed values. In machine learning, RMSD provides an advantage as it gives a clear sense of the average magnitude of prediction errors, making it a valuable tool for assessing the accuracy and reliability of regression models.

$$
RMSD = \sqrt{\frac{1}{N} \sum_{i=1}^{N} (y_i - \hat{y}_i)^2}
$$
Where:
- $N$ is the total number of observations.
- $y_i$ is the observed value for the i-th observation.
- $\hat{y}_i$ is the predicted value for the i-th observation.

```python
import pandas as pd
import numpy as np
import re

# Helper function to extract numerical value from a string
def extract_number(val):
    if isinstance(val, str):  # Check if the value is of type string
        # Extract the number from the string using regex
        match = re.search(r'(\d+\.?\d*)', val)
        # Return the number if a match is found, otherwise return NaN
        return float(match.group(1)) if match else np.nan
    return val  # If not string, return the original value

# List of specific columns to clean
columns_to_clean = ['back_legroom', 'front_legroom', 'fuel_tank_volume', 'highway_fuel_economy',
    'horsepower', 'interior_color', 'maximum_seating', 'mileage',
    'power', 'torque', 'length', 'height', 'maximum_seating', 'wheelbase', 'width']

# Clean the specified columns in train_data dataframe
for column in columns_to_clean:
    train_data[column] = train_data[column].apply(extract_number)
    train_data[column] = pd.to_numeric(train_data[column], errors='coerce')  # This will convert invalid parsing to NaN

# Handle NaN values by filling them with the column mean
for column in columns_to_clean:
    train_data[column].fillna(train_data[column].mean(), inplace=True)

# If you have a test_data dataframe and want to clean it in the same way:
for column in columns_to_clean:
    test_data[column] = test_data[column].apply(extract_number)
    test_data[column] = pd.to_numeric(test_data[column], errors='coerce')
    test_data[column].fillna(test_data[column].mean(), inplace=True)
```

```python
## Task 2, Question 3 Code Here

import pandas as pd

# Function to impute missing values
def impute_missing_values(df):
    for column in df.columns:
        if df[column].dtype in ['int64', 'float64']:  # If feature is numerical
            median_val = df[column].median()
            df[column].fillna(median_val, inplace=True)
        else:  # If feature is categorical or boolean
            mode_val = df[column].mode()[0]
            df[column].fillna(mode_val, inplace=True)
    return df

# Impute missing values in train_data
train_data = impute_missing_values(train_data)

# Impute missing values in test_data
test_data = impute_missing_values(test_data)
```

**Task 2, Question 4**: Encode all categorical variables appropriately as discussed in class.

- Where multiple values are given for an observation encode the observation as 'other'.
- Where a categorical feature contains more than 5 unique values, map the features into 5 most frequent values + 'other' and then encode appropriately. For instance, map colours into 5 basic colours + 'other': [red, yellow, green, blue, purple, other] and then encode.  
(2 marks)

```python
import pandas as pd

# Function to reduce categories and encode
def reduce_and_encode(data_train, data_test, categorical_columns):
    for col in categorical_columns:
        # Handle multiple values in an observation
        data_train[col] = data_train[col].apply(lambda x: 'other' if ',' in str(x) else x)
        data_test[col] = data_test[col].apply(lambda x: 'other' if ',' in str(x) else x)
        
        # Reduce categories to top 5 + 'other'
        if data_train[col].nunique() > 5:
            # Get the top 5 most frequent categories in the training data
            top_categories = data_train[col].value_counts().index[:5]
            
            # Replace categories not in top 5 with 'other' in both training and test data
            data_train[col] = data_train[col].apply(lambda x: x if x in top_categories else 'other')
            data_test[col] = data_test[col].apply(lambda x: x if x in top_categories else 'other')
            
    # Apply one-hot encoding
    data_train = pd.get_dummies(data_train, columns=categorical_columns, prefix=categorical_columns, drop_first=True)
    data_test = pd.get_dummies(data_test, columns=categorical_columns, prefix=categorical_columns, drop_first=True)
    
    # Align train and test data
    data_train, data_test = data_train.align(data_test, join='left', axis=1, fill_value=0)
    
    return data_train, data_test

# Identify categorical columns
categorical_cols = train_data.select_dtypes(include=['object']).columns

# Reduce categories and encode
train_data_encoded, test_data_encoded = reduce_and_encode(train_data, test_data, categorical_cols)

# Check the data
print(train_data_encoded.head())
print(test_data_encoded.head())
```

```python
from scipy.stats import ttest_ind

# Example: T-test between prices of new and used cars
new_car_prices = train_data_encoded[train_data_encoded['is_new'] == True]['price']
used_car_prices = train_data_encoded[train_data_encoded['is_new'] == False]['price']

t_stat, p_value = ttest_ind(new_car_prices, used_car_prices)
print('T-statistic:', t_stat)
print('P-value:', p_value)
```

T-statistic: 26.935809750898862

This is a measure of the difference between the two group means (new car prices and used car prices) in terms of the number of standard errors. A large absolute value for the t-statistic indicates a larger difference between the groups.

P-value: 2.2117893677458336e-145

The p-value represents the probability of observing the given t-statistic (or an even more extreme t-statistic) if the true difference between the group means is zero (null hypothesis). In this case, the p-value is extremely close to 0, which is much lower than the conventional significance level of 0.05.
Interpretation:
Given the large t-statistic and the extremely low p-value, we can reject the null hypothesis and conclude that there's a statistically significant difference between the prices of new cars and used cars. In simpler terms, new cars and used cars do not have the same average price; one is likely priced higher than the other.
